# Results

```python
import joblib
pd.set_option('display.max_rows', 10000)

results = pd.DataFrame(joblib.load("models/results.joblib"))
```

```python
results = results.sort_values("mean_test_score", ascending=False)
```

```python
def all_results(results):
    return results[["param_model", "rank_test_score", "mean_test_score", "mean_train_score"]]

def name(x):
    return  x: str(x)

def svc_results(results):
    results = results[results['param_model'].apply(name) == 'SVC'] \
        [["param_model", "mean_test_score", "mean_train_score",
            "param_model__C", "param_model__kernel", "param_model__gamma"]]
    results.columns = ["model",  "mean_test_score", "mean_train_score", "C", "kernel", "gamma"]
    return results

def knn_results(results):
    results = results[results['param_model'].apply(name) == 'KNeighborsClassifier'] \
        [["param_model", "mean_test_score", "mean_train_score",
            "param_model__n_neighbors", "param_model__weights"]]
    results.columns = ["model",  "mean_test_score", "mean_train_score", "n_neighbors", "weights"]
    return results
    
def tree_results(results):
    results = results[results['param_model'].apply(name) == 'RandomForestClassifier'] \
        [["param_model", "mean_test_score", "mean_train_score",
            "param_model__n_estimators", "param_model__criterion", "param_model__min_samples_split"]]
    results.columns = ["model",  "mean_test_score", "mean_train_score", 
                       "n_estimators", "criterion", "min_samples_split"]
    return results
```

```python
all_results(results)
```
